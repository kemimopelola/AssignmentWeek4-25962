```python
import pandas as pd
```

```python
homelessness = pd.read_csv('homelessness.csv')
```

```python
print(homelessness)
```

```python
homelessness["total"] = homelessness["individuals"] + homelessness["family_members"]
```

```python
print(homelessness)
```

```python
homelessness["indiv_per_10k"] = homelessness["individuals"] / homelessness["state_pop"] * 10000
```

```python
print(homelessness)
```

```python
high_homelessness = homelessness[homelessness["indiv_per_10k"] > 20]
```

```python
print(high_homelessness)
```

```python
high_homelessness_srt = high_homelessness.sort_values("indiv_per_10k", ascending = False)
```

```python
print(high_homelessness_srt)
```

```python
state_indiv_per_10k = high_homelessness_srt[["state","indiv_per_10k"]]
```

```python
print(state_indiv_per_10k)
```
